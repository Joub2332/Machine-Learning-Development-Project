### $\textbf{3. Feature Selection with box plots(optional)}$


For more precision in the feature selection, we can also plot box plots for each feature by class. This is a valuable tool for determining whether a feature provides useful predictive information and whether it can contribute to the model's performance.

````python
# Concatenate data and y to create a combined dataframe
data_combined = pd.concat([data, y], axis=1)
features = ['variance', 'skewness', 'curtosis', 'entropy']
for i, feature in enumerate(features, 1):
    #plt.subplot(2, 2, i)  # Crée un sous-graphe (2x2)
    data_combined.boxplot(column=feature, by='class')
    plt.title(f"Boxplot de {feature} par classe")
    plt.suptitle('')  # Supprime le titre par défaut généré par Matplotlib
    plt.xlabel("Classe")
    plt.ylabel(feature)

plt.show()
````

### Comment:

The boxplot analysis reveals that **variance** and **curtosis** are the most discriminative features for separating the classes. Variance shows a broader range for class 0, with visible outliers beyond 2 and -2, while class 1 has a narrower and more consistent distribution. Kurtosis also highlights clear differences, as class 0 exhibits a compact distribution centered around 0, whereas class 1 has a wider spread with notable outliers exceeding 3. **Skewness** provides some level of discrimination, as class 0 tends to have slightly higher asymmetry and broader variation than class 1. In contrast, **entropy** appears to have less discriminative power, with nearly identical distributions for both classes, showing significant overlap.

### Training Models with Cross-Validation

In this step, we implement a cross-validation pipeline to train multiple machine learning models and select the best-performing instance for each model type. The `training` function takes a set of predefined cross-validation splits, a dictionary of models, and the training data. For each model, it iterates through the folds, trains a fresh instance of the model on the training split, and evaluates it on the validation split. The validation accuracy is calculated for each fold, and the model instance with the highest accuracy is selected as the "best model" for that type. The result is a dictionary of the best-performing trained models, which ensures robust evaluation and comparison across different algorithms while minimizing the risk of overfitting.

## Good Programming Practices and Their Implementation in the Project

### 1. Code Modularity and Reusability
The project is designed with modular functions that perform specific tasks:
- **Ease of Maintenance**: Each function can be updated or debugged independently.
- **Reusability**: Functions can be reused with minimal changes for similar projects or datasets, you have to only modify the path to your dataset in the preparation function. 

### 2. Documentation and Comments
Every function is documented with detailed docstrings that describe:
- The function's purpose.
- Input parameters and expected types.
- Return values.
- Possible exceptions or errors.

This enhances code readability and helps new contributors understand the codebase quickly.

### 3. Error Handling and Assertions
The `preprocessing()` function uses `assert` statements to validate inputs and prevent unexpected issues:
```python
assert dataset == 'data_banknote_authentication.txt', "Invalid dataset"
```

### 4. Consistent Naming Conventions
The project follows **PEP 8** naming conventions:
- Functions and variables use `snake_case` for better readability.
- Descriptive naming ensures the code is easy to understand at a glance.

### 5. Separation of Concerns
The project applies the **Separation of Concerns** principle by organizing the code into distinct functions for different tasks:
- **`preprocessing()`**: Handles data cleaning and feature engineering.
- **`preparation()`**: Splits data into training and test sets and sets up cross-validation.
- **`training()`**: Trains various machine learning models.
- **`results()`**: Evaluates model performance.

This structure enhances the code's maintainability and facilitates testing and debugging.

### 6. Use of Libraries and Best Practices
The project utilizes libraries such as `pandas`, `numpy`, `scikit-learn`, and `matplotlib`:
- **`pandas` and `numpy`**: Efficient data manipulation.
- **`scikit-learn`**: For implementing machine learning models and evaluation metrics.
- **`matplotlib`**: Used for visualizing results.

The use of **K-Fold Cross-Validation** ensures robust model evaluation and reduces the risk of overfitting, providing a more reliable measure of model performance.

### 7. Preprocessing and Data Normalization
Customized data preprocessing ensures that each dataset is clean and suitable for training:
- Missing values are handled appropriately (e.g., filling with mean or mode).
- Data is centered and normalized to improve the stability and performance of models.

Example:
```python
# Centering data
data = (data - data.mean()) / data.std()
```


### 8. Cross-Validation for Robust Model Evaluation
The project uses **K-Fold Cross-Validation** to ensure robust model evaluation:
- This technique splits the training data into `k` subsets, training the model `k` times, each time using a different subset as the validation set.
- Provides more reliable performance metrics by testing the model's generalizability across different data subsets.

### 9. Reproducibility
Reproducibility is maintained by setting the `random_state` parameter:
```python
kf = KFold(nb_splits, shuffle=True, random_state=42)
```
### 10. Scalability
The code structure is designed for scalability:

- New models can be added by simply extending the models dictionary in the training() and results() functions.
- This makes the project flexible for future enhancements, including testing additional machine learning algorithms or working with new datasets.
